# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500  #necesitamos 500 datos generados

gender_data = []
income_data = []

#hacemos un bucle
for i in range(0,500):
    gender_data.append(np.random.choice(gender)) #para elegir aleatoriamente un elemento de gender
    income_data.append(np.random.choice(income)) #para elegir aleatoriamente un elemento de income
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
#si quiero más:
# para estilizar según distribución de media m y desviación tipica s es N(m,s) -> m + s * Z
# la Z en este curso siempre significa una Normal de media 0 y Std 1 Z -> N(0,1)
height = 160 + 30 * np.random.randn(n)  #la altura suele tener distribución normal
weight = 65 + 25 * np.random.randn(n)  #el peso suele tener distribución normal
age = 30 + 12 * np.random.randn(n)     # suele tener distribución 
income = 18000 + 3500 * np.random.randn(n)
```

```python
#genero el dataframe
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }


)
```

```python
data.head()
```

# Agrupación


```python
#agrupo por genero
grouped_gender = data.groupby("Gender") #crea un objeto con 2 atributos, uno de ellos es el nombre y el grupo

#Estos atributos se pueden acceder con sintaxis a traves de puntos
```

```python
grouped_gender.groups
```

```python
# hecha la agrupación, puede iterar sobre cada una de las opciones.  ej con un bucle
# util cuando hay muchos grupos y se necesitan operaciones separadas para cada grupo


for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
#genero otro grupo
grouped_Economic_Status = data.groupby("Economic Status")
```

```python
grouped_Economic_Status.groups
```

```python
grouped_Economic_Status.get_group("Rich")
```

```python
#para varios
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

```python
#los objetos groupby se comportan como cualquier dataframe. asi que podemos seleccionar una de esas columnas para realizar cualquier operación
grouped_income = double_group["Income"] 
#para poder ver primero una de las variables y leerla mejor.
```

```python
grouped_income.describe()
```

```python
#el metodo aggregate (groupby modificado) sirve para calcular conjuntamente los valores que nos interesen (suma de ingresos, media de edades, etc)
```

```python
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }

)
```

```python
#el máximo exponente de este tipo de agregados es la función lambda
#para definir un tipo de cálculo específico en python
#ej. en lugar de querer el promedio de la edad, queremos la tipificación osea...
#la media de la edad dividido por la desviación típica
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h:np.mean(h)/np.std(h)  #lo usual es llamarle x pero se puede como quiera, ej h
        # lambda es una función que se pasa como parámetro de la función aggregate, donde cualquier altura h se divide en la desviación std
    }

)
```

```python
#se pueden aplicar las funciones a todas las columnas
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
#para agregar con una lambbda, para aplicar funciones de transformación según haga falta
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Filtrado de datos

```python
double_group.sum()
```

```python
#aplico funcion filter y dentro una lambda
double_group["Age"].filter(lambda x: x.sum()>2500)  
#genera los elementos que pertenecen a un grupo cuya suma se mayor a 2500
```

```python
#hacer una transformacion matematica a los elementos de una categoria numerica
#ej calcular la tipificación de los valores de una columna en concreto (Zscore)
zscore = lambda x : (x - x.mean())/x.std() #a x le resto el promedio de todas las x y divido por la std de todas las x
```

```python
#el resultado es un dataframe donte tiene todos los datos normalizados o estandarizados
#para dibujar un plot después le colocamos nombre z_group

z_group = double_group.transform(zscore)
```

```python
plt.hist(z_group["Age"])
```

```python
#el secreto es el uso de la función transform que es útil en casos como cuando faltan datos.
fill_na_mean = lambda x : x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
